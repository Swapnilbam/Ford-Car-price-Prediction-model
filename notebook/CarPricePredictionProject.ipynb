## Creating Model : X_one_encoded

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import  train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
XO_train,XO_test,YO_train,YO_test = train_test_split(X_one_encoded,Y,test_size = 0.2,random_state = 42)
model = LinearRegression()
model.fit(XO_train,YO_train)
YO_pred = model.predict(XO_test)

MAE = mean_absolute_error(YO_test,YO_pred)
MSE = mean_squared_error(YO_test,YO_pred)
RMSE = np.sqrt(MSE)
R2 = r2_score(YO_test,YO_pred)
n = len(YO_test)
p = X_one_encoded.shape[1]
adj_r2 = 1 - ((1-R2)*(n - 1)/(n - p - 1))

print("Evaluation of One hot Encoded MODEL : \n")
print(f'Mean Absolute Error (MAE) : {MAE}')
print(f'Root Mean Squared  Error (RMSE) : {RMSE}')
print(f'R2 Score : {R2 * 100:.2f}%')
print(f'Adjusted R2 Score : {adj_r2 * 100:.2f}%')
```

## Creating Model for Label Encoding 

```python
X_label.shape
```

```python
XL_train,XL_test,YL_train,YL_test = train_test_split(X_label,Y,test_size = 0.2,random_state  = 42)
Label_model = LinearRegression()
Label_model.fit(XL_train,YL_train)
YL_pred = Label_model.predict(XL_test)

label_MAE  = mean_absolute_error(YL_test,YL_pred)
label_MSE = mean_squared_error(YL_test,YL_pred)
label_RMSE = np.sqrt(label_MSE)
label_R2 = r2_score(YL_test,YL_pred)
n = len(YL_test)
p = X_label.shape[1]
label_ad_r2  = 1 - ((1 - label_R2) * (n - 1)/ (n - p - 1))
print("Evaluation of Label Encoded Model : \n")
print(f'Mean Absolute Error (MAE) : {label_MAE}')
print(f'Root Mean Squared Error (MAE) : {label_RMSE}')
print(f'R2 Score : {label_R2 * 100:.2f} %')
print(f'Adjusted R2 Score : {label_ad_r2 * 100:.2f} %')
```

## As we can see from the Evaluation result of Regression models, we observed that One hot Encoding model is giving more Accurate result with R2 score of 84.6 % whereas Label encoded model is giving accuracy of 73.66% only, so we will prefer to use ONE-HOT encoded model for predictions.
